```python
data=pd.read_csv("/home/adriana/Projects/CNV/Validation/All_Validation_CNVs_annotated-220802.csv")
txs=pd.read_csv("/home/adriana/Projects/CNV/Validation/gCNV-validation/txs.tsv",sep='\t')
txs["Transcript"]=txs["Transcript"].str.strip()
```

```python
val_samples=data.loc[data['Sample'].str.contains("NV")]
```

```python
val_samples=val_samples.loc[val_samples['CSQ_Feature'].isin(txs["Transcript"].tolist())]
```

```python
def run_background(cell_value):

    highlight = 'background-color: darkorange;'
    default = ''

    if cell_value == 39:
        return 'background-color: darkorange;'
    elif cell_value == 42:
        return 'background-color: blue;'
    elif cell_value == 50:
        return 'background-color: red;'


#val_samples.style.applymap(run_background)
```

```python
from openpyxl import drawing, load_workbook
from openpyxl.styles import Alignment, Border, DEFAULT_FONT, Font, Side
from openpyxl.styles.fills import PatternFill
from openpyxl.utils import get_column_letter
```

```python
writer = pd.ExcelWriter('/home/adriana/Projects/CNV/Validation/CNV_Validation_Supplementary.xlsx', engine='openpyxl')
workbook= writer.book



def highlight(s):
    if s.Run == 39:
        return ['color:black;background-color: thistle'] * len(s)
    elif s.Run == 42:
        return ['color:black;background-color: lightsteelblue'] * len(s)
    elif s.Run == 50:
        return ['color:black;background-color: darkseagreen'] * len(s)
    else:
        return [''] * len(s)

val_samples.style.apply(highlight, axis=1).to_excel(writer,sheet_name='1 - Concordance',index=False)

```

```python
run39=val_samples[val_samples["Run"]==39]
run39=run39[run39.duplicated(subset=["ID"],keep=False)]

run42=val_samples[val_samples["Run"]==42]
run42=run42[run42.duplicated(subset=["ID"],keep=False)]
run50=val_samples[val_samples["Run"]==50]
run50=run50[run50.duplicated(subset=["ID"],keep=False)]

experiment3=pd.concat([run39,run42,run50],ignore_index=True)
experiment3.style.apply(highlight, axis=1).to_excel(writer,sheet_name='3 - KnownCNV',index=False)
```

```python

experiment4=experiment3.loc[experiment3["Run"]==39]
experiment4.style.apply(highlight, axis=1).to_excel(writer,sheet_name='4 - Repeatability',index=False)
```

```python
rep_samples=("CEN26","CEN32","CEN33","CEN35")


experiment5=val_samples.loc[val_samples["Sample"].str.startswith(rep_samples)]

experiment5.style.apply(highlight, axis=1).to_excel(writer,sheet_name='5 - Reproducibility',index=False)



```
